```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch
```

#GPT-2 is a popular model for text generation tasks.

```python
tokenizer_gpt2 = AutoTokenizer.from_pretrained("gpt2")
model_gpt2 = AutoModelForCausalLM.from_pretrained("gpt2")

generator_gpt2 = pipeline('text-generation', model=model_gpt2, tokenizer=tokenizer_gpt2)
prompt = "Once upon a time"
generated_text_gpt2 = generator_gpt2(prompt, max_length=50)
print(generated_text_gpt2[0]['generated_text'])
```

#A smaller, distilled version of GPT-2, which is faster and uses less memory.

```python
tokenizer_distilgpt2 = AutoTokenizer.from_pretrained("distilgpt2")
model_distilgpt2 = AutoModelForCausalLM.from_pretrained("distilgpt2")

generator_distilgpt2 = pipeline('text-generation', model=model_distilgpt2, tokenizer=tokenizer_distilgpt2)
generated_text_distilgpt2 = generator_distilgpt2(prompt, max_length=50)
print(generated_text_distilgpt2[0]['generated_text'])
```

#BERT is typically used for tasks like text classification, named entity recognition, etc., not text generation. However, it can be adapted for text generation using masked language modeling.

```python
from transformers import BertTokenizer, BertForMaskedLM
tokenizer_bert = BertTokenizer.from_pretrained('bert-base-uncased')
model_bert = BertForMaskedLM.from_pretrained('bert-base-uncased')

sentence = "[CLS] I love to [MASK] new things. [SEP]"
input_ids = tokenizer_bert.encode(sentence, return_tensors='pt')
mask_token_index = torch.where(input_ids == tokenizer_bert.mask_token_id)[1]

with torch.no_grad():
    output = model_bert(input_ids)
logits = output.logits

mask_token_logits = logits[0, mask_token_index, :]
top_5_tokens = torch.topk(mask_token_logits, 5, dim=1).indices[0].tolist()
for token in top_5_tokens:
    word = tokenizer_bert.decode([token])
    print(f"Prediction: {word}")
```

#T5 is a versatile model that treats every NLP problem as a text-to-text problem. It can be used for translation, summarization, and more.

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load T5 model and tokenizer
tokenizer_t5 = T5Tokenizer.from_pretrained("t5-small")
model_t5 = T5ForConditionalGeneration.from_pretrained("t5-small")

# Example task: Summarization
text = "The quick brown fox jumps over the lazy dog. The dog was not amused by the fox's antics."
input_text = "summarize: " + text
input_ids = tokenizer_t5.encode(input_text, return_tensors="pt")

# Generate summary
summary_ids = model_t5.generate(input_ids, max_length=30, min_length=5, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer_t5.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```

#BART is a powerful model for text generation tasks, including summarization and translation.

```python
from transformers import BartTokenizer, BartForConditionalGeneration

# Load BART model and tokenizer
tokenizer_bart = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model_bart = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

# Example task: Summarization
text = "Artificial intelligence is a field of study that focuses on the creation of intelligent agents capable of performing tasks that require human intelligence. It has applications in various domains such as healthcare, finance, and robotics."
input_ids = tokenizer_bart.encode(text, return_tensors="pt")

# Generate summary
summary_ids = model_bart.generate(input_ids, max_length=30, min_length=5, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer_bart.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```

#MarianMT is a family of models for machine translation.

This LLM is the by far I enjoyed trying :D


```python
from transformers import MarianMTModel, MarianTokenizer

# Load MarianMT model and tokenizer for English to French translation
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer_marian = MarianTokenizer.from_pretrained(model_name)
model_marian = MarianMTModel.from_pretrained(model_name)

# Example task: Translation
text = "I love programming and artificial intelligence."
input_ids = tokenizer_marian.encode(text, return_tensors="pt")

# Generate translation
translated_ids = model_marian.generate(input_ids, max_length=40)
translation = tokenizer_marian.decode(translated_ids[0], skip_special_tokens=True)
print(translation)
```

#DistilBERT is a smaller, faster version of BERT. It can be used for tasks like text classification.

```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from transformers import TextClassificationPipeline

# Load DistilBERT model and tokenizer
tokenizer_distilbert = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model_distilbert = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

# Initialize pipeline for text classification
classifier = TextClassificationPipeline(model=model_distilbert, tokenizer=tokenizer_distilbert)

# Example task: Sentiment analysis
text = "I am very happy with the service."
result = classifier(text)
print(result)
```

#ALBERT is a more parameter-efficient version of BERT.

```python
from transformers import AlbertTokenizer, AlbertForMaskedLM

# Load ALBERT model and tokenizer
tokenizer_albert = AlbertTokenizer.from_pretrained("albert-base-v2")
model_albert = AlbertForMaskedLM.from_pretrained("albert-base-v2")

# Example task: Masked Language Modeling
sentence = "[CLS] The capital of France is [MASK]. [SEP]"
input_ids = tokenizer_albert.encode(sentence, return_tensors='pt')
mask_token_index = torch.where(input_ids == tokenizer_albert.mask_token_id)[1]

with torch.no_grad():
    output = model_albert(input_ids)
logits = output.logits

mask_token_logits = logits[0, mask_token_index, :]
top_5_tokens = torch.topk(mask_token_logits, 5, dim=1).indices[0].tolist()
for token in top_5_tokens:
    word = tokenizer_albert.decode([token])
    print(f"Prediction: {word}")
```
